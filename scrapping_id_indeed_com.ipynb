```python
from bs4 import BeautifulSoup
import requests
import time
import os


pages = [10,20,30,40,50,60,70,80,90]

# with open(data.csv, 'a', encoding='utf-8', newline='') as f_output:
#     cav_print = cav.writer(f_output)
#     file_is_empty = os.stat(datascientist.csv).st_size == 0
#     if  file_is_empty:
#         csv_print.writerow(['Job_Title', 'Company', 'Location', 'Summary', 'Salary' ])
courses_list=[]    
for page in pages:
    source = requests.get('https://id.indeed.com/jobs?q=data+scientist&start={}'.format(page)).text
    soup = BeautifulSoup(source, 'lxml')
    
    for jobs in soup.find_all(class_='result'):
        try:
            jobs_title = jobs.a.text.strip()
        except Exception as e:
            job_title = None
        print("jobs_title : ", jobs_title)

        try:
            company = jobs.span.text.strip()
        except Exception as e:
            company = None
        print("company : ", company)

        try:
            location = jobs.find('span', class_='location').text.strip()
        except Exception as e:
            location = None
        print("location : ", location)

        try:
            summary = jobs.ul.text.strip()
        except Exception as e:
            summary = None
        print("summary : ", summary)

        try:
            salary = jobs.find('span', class_='salaryText').text.strip()
        except Exception as e:
            salary = None
        print("salary: ", salary)
#         csv_print.writeraw([job_title, company, location, summary, salary])
        course=[jobs_title,company,location, summary, salary]
        courses_list.append(course)

        print("===========================================")
        time.sleep(0.5)
        
```

```python
import csv

with open ('filename.cv','wb') as file:
    writer=csv.writer(file)
    for row in courses_list:
        writer.writerow(row)
```
