```python
from bs4 import BeautifulSoup

dokumen = '''
<html>
<head>
    <title>Tutorial BeautifulSoup</title>
</head>

<body>
    <p class="judul">Judul Dokumen</p>

    <p class="paragraf">Ini adalah contoh paragraf</p>

    <a href="https://ngodingdata.com" class="url">Ngodingdata</a>
</body>

</html>
'''

html = BeautifulSoup(dokumen, 'html.parser')


```

```python
judul = html.find('p')
judul
```

```python
paragraf = html.find('p', class_='paragraf')
```

```python
isi_judul = html.find('p',class_='judul').text
isi_judul
```

```python
all_paragraf = html.find_all('p')
all_paragraf
```

```python
import requests

# Set headers
headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})
```

```python
from bs4 import BeautifulSoup

...
url = "http://quotes.toscrape.com/"
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
title = soup.find('title').text
title
```

```python
sentence = soup.find('span', 'text' ).text
sentence
```

```python
author = soup.find_all('small', 'author')
author
```

```python
q = soup.find_all('span', class_='text')
q


    
```
